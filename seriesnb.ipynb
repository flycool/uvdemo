```python
import numpy as np
np.random.seed(42)
arr = np.random.randint(50, 101, 10)
print(arr)
s = pd.Series(arr, index=['student' + str(i) for i in range(1, 11)])
print(s)

print(s.mean())
print(s.max())
print(s.min())
print(s[s>73.7].size)
```

```python
t = pd.Series([28, 31, 29, 32, 30, 27, 33],
index=[1, 2, 3, 4, 5, 6, 7]
)
print(t[t>30].size)
print(t.mean())
print(t.sort_values(ascending=False))
print(t)

print(t.var())

t1 = t.diff()
nt1 = t1.drop_duplicates().abs()
snt1 = nt1.sort_values(ascending=False)

print(snt1.keys()[:2].tolist())
```

```python
s = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2], index=pd.date_range('2023-01-01', periods=10))
print(s)

s1 = s.pct_change()
print(s1)

print(s.idxmax)
print(s.idxmin)
# print(s.keys()[s.argmax()])
# print(s.keys()[s.argmin()])
print(s1.std())
```

```python
s = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220], index=pd.date_range('2022-01-01', periods=12, freq='MS'))

s1 = s.resample('QS').mean()
print(s1)
print(s.idxmax())
a = s.pct_change()
b = a>0
c = b[b.rolling(3).sum()==3]
print(c.keys())



```
