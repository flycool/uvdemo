```python
df = pd.DataFrame({
    '姓名': ['张三', '李四', '王五', '赵六', '钱七'],
    '数学': [85, 92, 78, 88, 95],
    '英语': [90, 88, 85, 92, 80],
    '物理': [75, 80, 88, 85, 90]
})

df
```

```python
df['total'] = df[['数学', '英语', '物理']].sum(axis=1)
df['avg'] = df[['数学', '英语', '物理']].mean(axis=1)

df[(df['数学']>90) | (df['英语']>85)]

# df.sort_values('total',ascending=False).head(3)

df.nlargest(3, columns='total')
```

```python
data = {
    '产品名称': ['A', 'B', 'C', 'D'],
    '单价': [100, 150, 200, 120],
    '销量': [50, 30, 20, 40]
}
df = pd.DataFrame(data)
df
```

```python
df['total'] = df['单价'] * df['销量']
df

df.nlargest(1, columns='total')

df.sort_values('total', ascending=False)
```

```python
data = {
    '用户ID': [101, 102, 103, 104, 105],
    '用户名': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    '商品类别': ['电子产品', '服饰', '电子产品', '家居', '服饰'],
    '商品单价': [1200, 300, 800, 150, 200],
    '购买数量': [1, 3, 2, 5, 4]
}
df = pd.DataFrame(data)
df
```

```python
df['total'] = df['商品单价'] * df['购买数量']
df
```

```python
df.nlargest(1, columns='total')
df['total'].mean()
```
