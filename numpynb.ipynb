```python
import numpy as np
```

```python
arr = np.array([1,2,3])
print(arr)
```

```python
arr = np.logspace(0, 4, 3, base=2)
print(arr)
```

```python
a = np.exp(2)
print(a)
```

```python
arr = np.array([28, 30, 29, 31, 32, 30, 29])
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
print(arr[arr>30].size)
```

```python
arr = np.array([85, 90, 78, 92, 88])
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(arr/10)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(a+b)
print(a*b)
print(a@b)
```

```python
np.random.seed(0)
arr = np.random.randint(0, 10, (3, 4))
print(arr)
print(np.max(arr, axis=0))
print(np.min(arr, axis=1))
print(np.where(arr%2!=0, -1, arr))
arr[arr%2==1] = -1
print(arr)
```

```python
arr = np.arange(1, 13)
print(arr)
arr = np.reshape(arr, (3, 4))
print(arr)
print(np.sum(arr, axis=1))
print(np.mean(arr, axis=0))
print(np.reshape(arr, 12))
```

```python
np.random.seed(0)
arr = np.random.randint(0, 20, (5, 5))
print(arr)
print(arr[arr>10])
arr[arr>10] = 0
print(arr)
```

```python
arr = np.array([120, 135, 110, 125, 130, 140])
print(f"{np.sum(arr)}")
print(f"{np.mean(arr):.3f}")
print(f"{np.var(arr):.3f}")
print(np.argmax(arr)+1)
print(np.argmin(arr)+1)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.concatenate([a, b])
print(c)
print(np.reshape(c, (2,3)))
```

```python
arr = np.array([2, 1 ,2, 3, 1, 4, 3])
uarr, counts = np.unique(arr, return_counts=True)
print(uarr)
print(counts)
```
